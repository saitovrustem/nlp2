Импортируем модуль с логистической регрессией (ей мы и будем классифицировать тексты) и с подсчетом по тексту частот слов, входящих в него (CountVectorizer). Сразу считаем всю выборку, чего, конечно, нельзя ни в коем случае делать, если она велика.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer

input_file = open(data_path, 'r')
text = input_file.read()
```

Если бы текстовый файл был очень большой, нас бы спас генератор input_file.xreadlines(), чтобы не грузить его целиком в память. Также будет полезно изучить пример классификации данных не помещающихся в память из документации sklearn: http://scikit-learn.org/stable/auto_examples/applications/plot_out_of_core_classification.html#example-applications-plot-out-of-core-classification-py

Тем временем разобьем строку на список строк по символу переводf строки и посмотрим несколько первых строк, чтобы убедиться, что все хорошо:

```python
print "\n".join(text.split("\n")[:3])
```

Похоже, что все в порядке. Также внешне очень похоже, что разделителем для метки класса и текста служит табуляция. Проверим:

```python
print (text.split("\n")[0].split('\t'))
```

Все верно. Тогда понятно, как можно получить вектор с ответами на обучающей выборке. Изначально в коде ниже не было условия на ненулевую длину строки, но оно добавлено из-за перевода строки в конце файла. Также не было замены меток "spam" и "ham" на 1 и 0 и приведения к int. Однако ниже мы будем использовать для оценки качества f1-меру и стандартная функция из библиотеки sklearn попросит нас использовать такие обозначения. Что ж, нам не трудно.

```python
y = [int(line.split('\t')[0].replace('ham', '0').replace('spam','1')) for line in text.split('\n') if len(line) > 0]
```

```python
print y[:5]
```

Аналогично создаем список текстов смс-сообщений:

```python
texts = [line.split('\t')[1] for line in text.split('\n') if len(line) > 0]
```

```python
print len(texts)
```

Теперь создадим экземпляр класса CountVectorizer и вызовем у него метод fit_transform, чтобы сразу и обучить словарь и вернуть матрицу признаков для текстов. Есть и раздельные методы fit() и transform() 

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
```

Интересно, какого типа полученная матрица. Ясно, что в одном тексте используется лишь небольшая часть словаря. Значит, в матрице будет много нулей. Хранить кучу нулей в памяти выглядит довольно странно, поэтому матрица записана в специальном формате для разреженных матриц.

```python
print type(X)
```

Давайте посмотрим на саму матрицу (библиотека scipy заботливо убережет нас от просмотра всех ее элементов, но покажет, как примерно выглядит X)

```python
print X
```

Это матрица в формате csr. Другие примеры разреженных матриц - csc и lil_matrix. С разреженным матрицами можно делать многие обычные матричные операции. Например, можно взвесить признаки, домножив матрицу признаков на диагональную. Ее в этом случае тоже лучше конструировать в виде csr матрицы. У разреженных матриц есть метод .toarray(), конвертирующий их в плотные. Никогда не используйте его, если уже разреженная матрица требует много памяти - ни к чему хорошему это привести не может. Но если плотная матрица поместиться в память может, а нужный алгоритм не реализован для разреженных матриц, то можно попробовать привести матрицу к плотному виду. Обращение по индексу в разреженной матрице тоже делается просто: одним из двух способов ниже. 

```python
print X[576, 1]
```

```python
print X[(576, 1)]
```

А строка достается так:

```python
print X[1]
```

Как можно заметить, полученная строка тоже матрица, просто из одной строки. Поэтому такое обращение не сработает:

```python
print X[1][4342]
```

```python
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

classifier = LogisticRegression() #LinearSVC() #MultinomialNB()
print np.mean(cross_val_score(classifier, X, np.array(y), scoring="f1"))
```
