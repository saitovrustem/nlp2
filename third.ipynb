```python
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
filenames = ['C:/project/data/austen-bronte/Austen_Emma.txt',
            'C:/project/data/austen-bronte/Austen_Pride.txt',
            'C:/project/data/austen-bronte/Austen_Sense.txt',
            'C:/project/data/austen-bronte/CBronte_Jane.txt',
            'C:/project/data/austen-bronte/CBronte_Professor.txt',
            'C:/project/data/austen-bronte/CBronte_Villette.txt']
vectorizer = CountVectorizer(input='filename')
dtm = vectorizer.fit_transform(filenames)  # a sparse matrix
vocab = vectorizer.get_feature_names()  # a list
```

```python
n, _ = dtm.shape
dist = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        x, y = dtm[i, :], dtm[j, :]
        dist[i, j] = np.sqrt(np.sum((x - y)**2))
from sklearn.metrics.pairwise import euclidean_distances
dist = euclidean_distances(dtm)
print np.round(dist, 1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
dist = 1 - cosine_similarity(dtm)
print np.round(dist, 2)
print dist[1, 3]
print dist[1, 3] > dist[3, 5]
```

```python
norms = np.sqrt(np.sum(dtm * dtm, axis=1, keepdims=True))  # multiplication between arrays is element-wise
dtm_normed = dtm / norms
similarities = np.dot(dtm_normed, dtm_normed.T)
print np.round(similarities, 2)
print similarities[1, 3]
```

```python
%matplotlib inline
import os  # for os.path.basename
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

mds = MDS(n_components=2, dissimilarity="precomputed", random_state=1)
pos = mds.fit_transform(dist)  # shape (n_components, n_samples)

xs, ys = pos[:, 0], pos[:, 1]
names = [os.path.basename(fn).replace('.txt', '') for fn in filenames]

for x, y, name in zip(xs, ys, names):
     color = 'orange' if "Austen" in name else 'skyblue'
     plt.scatter(x, y, c=color)
     plt.text(x, y, name)
        
plt.show()
```

```python
mds = MDS(n_components=3, dissimilarity="precomputed", random_state=1)
pos = mds.fit_transform(dist)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pos[:, 0], pos[:, 1], pos[:, 2])
for x, y, z, s in zip(pos[:, 0], pos[:, 1], pos[:, 2], names):
    ax.text(x, y, z, s)
        
plt.show()
```

```python
from scipy.cluster.hierarchy import ward, dendrogram
linkage_matrix = ward(dist)

print dendrogram(linkage_matrix, orientation="right", labels=names)

plt.tight_layout()
plt.show()
```
